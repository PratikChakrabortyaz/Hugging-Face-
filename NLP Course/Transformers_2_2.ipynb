```python
from transformers import AutoTokenizer
model1=AutoTokenizer.from_pretrained("bert-base-cased")
model2=AutoTokenizer.from_pretrained("roberta-base")
model3=AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
sentence="The quick brown fox jumps over the lazy dog"
tokens1=model1.tokenize(sentence)
tokens2=model2.tokenize(sentence)
tokens3=model3.tokenize(sentence)
print(f"Tokens1: {tokens1}")
print(f"Tokens2: {tokens2}")
print(f"Tokens3: {tokens3}")
```

```python
ids1=model1.convert_tokens_to_ids(tokens1)
ids2=model2.convert_tokens_to_ids(tokens2)
ids3=model3.convert_tokens_to_ids(tokens3)
print(f"Ids1: {ids1}")
print(f"Ids2: {ids2}")
print(f"Ids3: {ids3}")
```

```python
og_sentence1=model1.decode(ids1)
og_sentence2=model2.decode(ids2)
og_sentence3=model3.decode(ids3)
print(f"Original Sentence1: {og_sentence1}")
print(f"Original Sentence2: {og_sentence2}")
print(f"Original Sentence3: {og_sentence3}")
```
